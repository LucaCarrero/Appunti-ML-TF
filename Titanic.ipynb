```
import pandas as pd
import numpy as np

data = pd.read_csv('train.csv')
data.head()
data.replace({np.nan:'S'})
```

```
data = data.drop(columns=['PassengerId', 'Name', 'Ticket','Cabin'])
```

```
df=pd.get_dummies(data)

df = df.drop(columns=['Sex_male','Embarked_C'])
df = df.dropna()
df
```

```
X = df.drop(columns=['Survived'])
Y = df['Survived']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    Y, test_size=0.30, 
                                                    random_state=101)
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)

X_train = X_train.to_numpy()
X_test = X_test.to_numpy()

X_scaledTrain = scaler.transform(X_train)
X_scaledTest =  scaler.transform(X_test)

X_scaledTrain
```

#skearn

```
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
model = logmodel.fit(X_scaledTrain ,y_train)
```

```
predictions = logmodel.predict(X_test)
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

# keras

```
import keras #ML Library
import numpy as np #Mathematics Library
import pandas as pd #Data-handling Library
import matplotlib.pyplot as plt #Used to graph our results
import tensorflow as tf
```

```
inputs = tf.keras.Input(shape=(8,))
x = tf.keras.layers.Dense(4, activation=tf.nn.relu)(inputs)
outputs = tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)(x)
model = tf.keras.Model(inputs=inputs, outputs=outputs)
model.compile(keras.optimizers.Adam(lr=1), 'mean_squared_error')
```

```
history = model.fit(X_scaledTrain, y_train, epochs=200, batch_size=10)

plt.plot(history.history['loss'],'r' )
```

```
model.evaluate(X_scaledTest,y_test, batch_size=5)
```
